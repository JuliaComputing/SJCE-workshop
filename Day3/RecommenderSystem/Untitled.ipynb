```julia
using Images, ImageMagick
#using TSVD
using Interact

x = load("./images/strang.jpg")
img = map(x->Float64(x.r), x)
@manipulate for i=1:100
    LR = svds(img, nsv=i)
    U = LR[1][:U]
    S = LR[1][:S]
    V = LR[1][:Vt]
    map(U * (diagm(S) * V)) do a
        RGB(a, a, a)
    end
end
```
